```python
# Wines with score 99 or higher
topWines = reviews.loc[(reviews.points >= 99)]

# topWines
```

```python
# Sort wine by price
topWines.sort_values('price', ascending=False)
```

```python
#Get the most expensive wine
mostExpensiveWines = topWines.loc[(topWines.price == topWines.price.max())]

mostExpensiveWines
```

```python
# Filter wines that conain "cherry" in the description
pd.set_option('display.max_colwidth', None)
cherryWines = reviews.loc[reviews.description.str.contains('cherry', na=False)]

print(cherryWines.description)
```
