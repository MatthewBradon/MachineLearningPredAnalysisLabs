```python
import os
import pandas as pd
```

```python

pd.set_option('display.max_columns', None)

# na_values='?' is used to replace '?' with NaN
data = pd.read_csv('CSVs-Data/dataset_lab3.csv', header=None, na_values='?')

data.head()
```

```python
with open('CSVs-Data/feature_names_lab3.txt') as f:
    feature_names = f.read().splitlines()

data.columns = feature_names


data.head()
```

```python
# Print toal number of values, min, max and standard deviation of normalized-losses
print('count', data['normalized-losses'].count())
print('mean', data['normalized-losses'].mean())
print('std', data['normalized-losses'].std())
print('min', data['normalized-losses'].min())
print('25%', data['normalized-losses'].quantile(0.25))
print('50%', data['normalized-losses'].quantile(0.50))
print('75%', data['normalized-losses'].quantile(0.75))
print('max', data['normalized-losses'].max())
print('Name:', data['normalized-losses'].name, 'dtype:', data['normalized-losses'].dtype)


data['normalized-losses'].describe()

```

```python
print(data['make'].unique())

print('##########')

print(data['make'].mode())

print('##########')

print(data['make'].value_counts())
```

```python
# Create dataframe row report of unormalized-losses feature
row_report = pd.DataFrame(columns=['Feature', 'Count', '%Miss', 'Card', 'Min', '1stQrt', 'Mean', 'Median', '3rdQrt', 'Max', 'Std. Dev'])


# Create a row for normalized-losses feature
row = {
    'Feature': 'normalized-losses',
    'Count': data['normalized-losses'].count(),
    '%Miss': data['normalized-losses'].isnull().mean(),
    'Card': data['normalized-losses'].nunique(),
    'Min': data['normalized-losses'].min(),
    '1stQrt': data['normalized-losses'].quantile(0.25),
    'Mean': data['normalized-losses'].mean(),
    'Median': data['normalized-losses'].median(),
    '3rdQrt': data['normalized-losses'].quantile(0.75),
    'Max': data['normalized-losses'].max(),
    'Std. Dev': data['normalized-losses'].std()
}

# Convert row to pandas series
row = pd.Series(row)

# loc is used to add a row at the end of the dataframe
row_report.loc[len(row_report.index)] = row

row_report
```

```python
# Create row report for all features

row_report = pd.DataFrame(columns=['Feature', 'Count', '%Miss', 'Card', 'Min', '1stQrt', 'Mean', 'Median', '3rdQrt', 'Max', 'Std. Dev'])

print(data.columns)

for feature in data.columns:

    # Check if feature is categorical
    if data[feature].dtype == 'object':
       continue
    else:
        row = {
            'Feature': feature,
            'Count': data[feature].count(),
            '%Miss': data[feature].isnull().mean(),
            'Card': data[feature].nunique(),
            'Min': data[feature].min(),
            '1stQrt': data[feature].quantile(0.25),
            'Mean': data[feature].mean(),
            'Median': data[feature].median(),
            '3rdQrt': data[feature].quantile(0.75),
            'Max': data[feature].max(),
            'Std. Dev': data[feature].std()
        }
    row = pd.Series(row)
    row_report.loc[len(row_report.index)] = row

row_report

```
